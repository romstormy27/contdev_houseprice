# Menambahkan Grid ke Plot

- Terkadang, kita lebih memilih untuk menampilkan grid untuk mempermudah mencari poin referensi daripada latar belakang plot yang kosong.


- Dalam Matplotlib, grid bisa ditambahkan dengan menggunakan `plt.grid()`. Parameter yang digunakan adalah value Boolean: `True` untuk menambahkan grid, `False` untuk menghilangkan grid.

- Contoh di bawah menunjukkan multiline plot dengan data mock-up kasus positif penyakit dari bulan Maret hingga Juni untuk setiap kelompok umur, seperti yang ada di notebook sebelumnya.

```python
# list value untuk x-axis
x = ['Maret', 'April', 'Mei', 'Juni']

# list value untuk y-axis
y1 = [0, 10, 300, 400]  # grup umur 0-17 tahun
y2 = [4,  340, 400, 600] # grup umur 18-59 tahun
y3 = [1, 150, 50, 500] # grup umur >= 60 tahun

# Gambarkan line plot
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tambahkan grid
plt.grid(True)

# Tunjukkan plot yang telah digambar
plt.show()
```

# Mengubah Axes

- Ada saatnya dimana dibutuhkan untuk mengubah-ubah limit dari sebuah plot untuk membantu membawa informasi dari plot dengan lebih baik.


- Untuk mengubah axis, dapat menggunakan `plt.axis()` dan memasukkan parameter list dengan urutan sebagai berikut:
 [x-min, x-max, y-min, y-max]


- Dengan menggunakan contoh yang sama dengan plot di atas, berikut contoh mengubah axis dalam plot:

```python
# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tentukan nilai axis x dan y: [x-min, x-max, y-min, y-max]
plt.axis([-1, 4, -100, 700])

# Tunjukkan plot yang telah digambar
plt.show()
```

- Pada contoh di atas, x-axis memiliki titik-titik yang terlihat sebagai string, namun dapat juga dibaca secara numerik mulai dari 0 hingga 3 (karena python memulai perhitungan dari 0). Sehingga pada titik yang terlihat sebagai `Maret` dapat dibaca sebagai titik pertama atau titik 0 dan seterusnya.

- Ada kalanya hanya salah satu axis yang butuh diubah-ubah.


- Matplotlib memberikan kontrol yang lebih leluasa dengan fungsi `plt.xlim()` untuk x-axis dan `plt.ylim()` untuk y-axis. Gunakan parameter [x-min, x-max] atau [y-min, y-max]

```python
# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tentukan limit x-axis
plt.xlim([0, 3])

# Tentukan limit y-axis
plt.ylim([0, 700])

# Tunjukkan plot yang telah digambar
plt.show()
```

# Mengubah X dan Y tick

- Tick adalah titik-titik referensi yuang ada di x-axis dan y-axis.


- Matplotlib memberikan opsi untuk mengubah-ubah tick, yang juga berpengaruh pada letak grid.


- Untuk mengubah tick, gunakan `plt.xticks()` untuk x-axis dan `plt.ytics()` untuk y-axis.

```python
# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tentukan tick y-axis
plt.yticks([0, 250, 500, 750])

# Tunjukkan plot yang telah digambar
plt.show()
```

# Menambahkan Label

- Label adalah salah satu elemen penting dari sebuah plot. Label akan sangat membantu pembaca untuk cepat mengerti data apa yang sedang mereka baca.


- Untuk menampilkan label, gunakan `plt.xlabel()` untuk x-axis dan `plt.ylabel()` untuk y-axis dengan memberikan parameter berupa string sesuai label yang cocok.

```python
# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tampilkan label untuk x-axis
plt.xlabel('Bulan')

# Tampilkan label untuk y-axis
plt.ylabel('Kasus positif')

# Tunjukkan plot yang telah digambar
plt.show()
```

# Menambahkan Judul

- Untuk menambahkan judul pada plot, gunakan `plt.title()` dengan memberikan parameter berupa string sesuai judul yang cocok.

```python
# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tampilkan judul
plt.title('Kasus Positif Penyakit tiap Bulan [mock-up]')

# Tunjukkan plot yang telah digambar
plt.show()
```

# Menambahkan Legenda

- Pada beberapa multiline plot di atas, ada 3 buah line di dalam satu plot. Selain dari pembuat plot, orang lain mungkin akan kesulitan untuk membedakan line yang mana yang merepresentasikan kategori yang mana. Legenda dapat membantu memberikan informasi tambahan untuk menjelaskan plot dengan lebih baik.


- Untuk menambahkan legenda, gunakan `ax.legend()`

- Gunakan parameter `loc` untuk menentukan letak dari legenda

Daftar letak legenda:
- ax.legend(loc=0) # let Matplotlib decide the optimal location

- ax.legend(loc=1) # upper right corner

- ax.legend(loc=2) # upper left corner

- ax.legend(loc=3) # lower left corner

- ax.legend(loc=4) # lower right corner

- ax.legend(loc=5) # right

- ax.legend(loc=6) # center left

- ax.legend(loc=7) # center right

- ax.legend(loc=8) # lower center

- ax.legend(loc=9) # upper center

- ax.legend(loc=10) # center

```python
fig, ax = plt.subplots()

# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Tampilkan legenda
ax.legend(['0-17 tahun','18-59 tahun','>= 60 tahun'], loc=2)

# Tunjukkan plot yang telah digambar
plt.show()
```

- Menampilkan legenda juga dapat dengan menggunakan parameter label di dalam `plt.plot()`

```python
fig, ax = plt.subplots()

# Gambarkan line plot dari variabel-variabel yang sudah ditentukan di atas
plt.plot(x, y1, label='0-17 tahun')
plt.plot(x, y2, label='18-59 tahun')
plt.plot(x, y3, label='>= 60 tahun')

# Tampilkan legenda
ax.legend()

# Tunjukkan plot yang telah digambar
plt.show()
```

# Object-Oriented  API

- Matplotlib menyediakan opsi untuk memberikan kontrol lebih banyak kepada user yang butuh untuk membuat plot yang lebih kompleks berupa Object-Oriented interface.

- Jika seperti yang sebelumnya user menggambarkan plot pada figure yang aktif, Object-Oriented API memerlukan figure dan axes yang spesifik.

- Contoh di bawah menunjukkan pembuatan 2 buah plot pada satu figure dengan 2 axes yang berbeda.

```python
# value x dari 0 hingga 10 sebanyak 100 data point
x = np.linspace(0, 10, 100)

# Buat 1 figure dengan 2 obyek axes menggunakan plt.subplots()
fig, ax = plt.subplots(2)


# Gambarkan plot pada axes pertama
ax[0].plot(x, np.sin(x))

# Gambarkan plot pada axes pertama
ax[1].plot(x, np.exp(x))


# Tunjukkan plot yang telah digambar
plt.show()
```

# Figure dan Subplots

- Untuk membuat subplots, pertama-tama gunakan `plt.figure()` untuk membuat figure baru

```python
fig = plt.figure()
```

- di figure yang sudah dibuat, bisa ditambahkan subplot pertama dari 4 subplot seperti berikut:

```python
# Membuat subplot pertama dari 4 subplot (2 row, 2 kolom, bagian 1)
ax1 = fig.add_subplot(2,2,1)
```

```python
# Membuat subplot ke 2
ax2 = fig.add_subplot(2,2,2)

# Membuat subplot ke 2
ax3 = fig.add_subplot(2,2,3)

# Membuat subplot ke 2
ax4 = fig.add_subplot(2,2,4)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# Tunjukkan plot yang telah digambar
plt.show()
```
