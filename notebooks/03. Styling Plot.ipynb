# Import Package

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import seaborn sebagai sns
import seaborn as sns

# Gunakan tema default
sns.set_theme()
```

# Load dataset

```python
tips = sns.load_dataset("tips")
tips.head()
```

# Multiplot grids

- Untuk mempermudah melihat dataset dengan cepat, `joinplot()` dan `pairplot()` dapat menjadi sangat membantu.

- `joinplot()` memperlihatkan scatterplot beserta distribusi dari tiap variabel.


- `pairplot()` memperlihatkan scatterplot dan distribution plot.


- Lihat dokumentasi untuk cara-cara lain dalam membuat multiplot grid.


Dokumentasi: https://seaborn.pydata.org/tutorial/axis_grids.html

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.pairplot(data=tips, hue="smoker")
```

# Seaborn Figure Style

Ada 5 tema preset seaborn, yaitu:
    1. darkgrid
    2. whitegrid
    3. dark
    4. white
    5. ticks


- Untuk mengubah tema plot seaborn, gunakan `seaborn.set_style()`

Dokumentasi: https://seaborn.pydata.org/tutorial/aesthetics.html

```python
sns.set_style("whitegrid")
sns.catplot(data=tips, 
            kind="box", 
            x="day", 
            y="total_bill"
           )
```

- Jika ingin mengubah style figure sementara, gunakan fungsi `axes_style()` dalam with statement. Ini juga membantu untuk menggambarkan beberapa plot dengan style yang berbeda dalam subplot.

```python
f = plt.figure(figsize=(8, 8))
# gridspec = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(2,2,1)
    sns.boxplot(data=tips, x="day", y="total_bill", ax=ax)

with sns.axes_style("white"):
    ax = f.add_subplot(2,2,2)
    sns.boxplot(data=tips, x="day", y="total_bill", ax=ax)

with sns.axes_style("ticks"):
    ax = f.add_subplot(2,2,3)
    sns.boxplot(data=tips, x="day", y="total_bill", ax=ax)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(2,2,4)
    sns.boxplot(data=tips, x="day", y="total_bill", ax=ax)

f.tight_layout()
```

# Menghilangkan Spines

- Untuk menghilangkan spine, gunakan fungsi `despine()`

```python
f = plt.figure(figsize=(8, 8))

sns.set_style("white")
sns.boxplot(data=tips, x="day", y="total_bill")
sns.despine()
```

- Fungsi `despine()` juga menyediakan parameter untuk menjauhkan spine dari data. Gunakan parameter `offset` untuk menyesuaikan spine dari plot.

```python
f = plt.figure(figsize=(8, 8))

sns.boxplot(data=tips, x="day", y="total_bill")
sns.despine(offset=20, trim=True);
```

Quiz:
- Gunakan `despine()` untuk menghilangkan spine di kiri, bawah, kanan, atau atas sesuai dengan keinginan.

```python
# Quiz
sns.set_style("whitegrid")

f = plt.figure(figsize=(8, 8))


sns.boxplot(data=tips, x="day", y="total_bill")
sns.despine(left=False, bottom=False, right=False, top=False)
```

# Override Elemen dari seaborn style

- seaborn menyediakan cara untuk kustomisasi style seaborn. Gunakan dictionary parameter rc argument dari `axes_style()` dan `set-style()`

```python
# Cek parameter yang dapat diubah
sns.axes_style()
```

```python
f = plt.figure(figsize=(8, 8))

sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.boxplot(data=tips, x="day", y="total_bill")
```
